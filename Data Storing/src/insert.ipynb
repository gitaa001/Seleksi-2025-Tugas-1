```python
import json
import psycopg2

# IMPORT TRAINEE

# Koneksi ke database
conn = psycopg2.connect(
    dbname="Produce48",
    user="postgres",
    password="Masukaja#224",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

# Load data trainee
with open('../../Data Scraping/data/cleaned/trainee.json', 'r', encoding='utf-8') as f:
    trainees = json.load(f)

for t in trainees:
    cur.execute("""
        INSERT INTO trainee (id_trainee, name, company, age, korean_name, japanese_name, first_grade, last_grade, first_rank, final_rank)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        ON CONFLICT (id_trainee) DO NOTHING
    """, (
        t.get('id_trainee'),
        t.get('name'),
        t.get('company'),
        t.get('age'),
        t.get('korean_name'),
        t.get('japanese_name'),
        t.get('1st_grade'),
        t.get('last_grade'),
        t.get('1st_rank'),
        t.get('final_rank')
    ))

conn.commit()
cur.close()
conn.close()
```
